```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
a = np.array([10,20])
b = np.array([8,19])
plt.plot(a,b,color='pink')
plt.show()
```

```python
x=[6,7,8,9,10]
y=[6,8,18,19,20]
plt.figure(figsize=(8,5))
plt.plot(x,y,color='pink',linestyle='-.',marker='*',linewidth=2)
plt.xlabel('Krupa')
plt.ylabel('Rupa')
plt.title("Statistics of Greatness")
plt.legend(['Starlight'])
plt.grid(False)
plt.show()
```

```python
categories=['Rupa','Krupa','Bindu','Deekshitha','Cherry','Chandhana','Pavani','Sindhu']
values=[10,9,9,8,5,6,4,3]
plt.figure(figsize=(8,5))
plt.bar(categories,values,color='pink')
plt.xlabel('Friends')
plt.ylabel('Score')
plt.title('Leaderboard')
plt.show()
```

```python
data = [12, 15, 13, 17, 16, 16, 19, 12, 14, 16, 14, 17, 18, 18, 15, 14]

plt.hist(data, bins=16, edgecolor='green',color='pink')
plt.xlabel('Rating')
plt.ylabel('BL Name')
plt.title('Historical BL')
plt.show()
```

```python
x = [38,50,67,90,78]
y = [33,40,76,91,56]

plt.scatter(x, y,color='purple',marker='*')
plt.xlabel('Birthdayy')
plt.ylabel('Timee')
plt.title('Happy Birthday!!')
plt.show()
```

```python
labels = ['cherry', 'Bindu', 'Pavani', 'Rupa']
sizes = [5,5,5,80]
colors=['lightyellow','lavender','lightblue','pink']
plt.pie(sizes,colors=colors, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Greatest Person')
plt.axis('equal')
plt.show()
```

```python
data = [7, 8, 5, 6, 7, 9, 12, 5, 7, 8, 6]

plt.boxplot(data)
plt.title('Cylinder')
plt.ylabel('Points')
plt.show()
```

```python
data = np.random.rand(10, 10)

plt.imshow(data,cmap='Pastel1', interpolation='nearest')
plt.colorbar()
plt.title('Simple Heatmap Example')
plt.show()
```
